**Task-04
To analyze distributions and frequency of variables.**

2. Plot histogram for age.
3. Plot histogram for cholesterol levels.
4. Interpret distributions.

Conclusion:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Heart_Disease_Prediction.csv")
```

```python
#1Count number of patients with and without heart disease.
heart_disease_counts = df['Heart Disease'].value_counts()
print("Number of Patients with Heart Disease:")
print(heart_disease_counts)
```

```python
#2Plot histogram for age.
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
#3Plot histogram for cholesterol levels.
plt.figure(figsize=(10, 6))
plt.hist(df['Cholesterol'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.title('Histogram of Cholesterol Levels')
plt.show()
```

**Task-05
To study relationships between variables using basic exploration.**

Tasks:
1. Compute correlation matrix.
2. Identify highly correlated variables.
3. Plot scatter plot between age and maximum heart rate.
4. Interpret relationships.

Conclusion:
Relationships between medical parameters were explored.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Heart_Disease_Prediction.csv")

```

```python
#1Compute correlation matrix.
correlation_matrix = df.corr(numeric_only=True)
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
#2 identify highly correlated variables
high_corr_threshold = 0.5
high_corr_variables = correlation_matrix[
    (correlation_matrix > high_corr_threshold) |
    (correlation_matrix < -high_corr_threshold)
].stack().reset_index()
high_corr_variables.columns = ['Variable 1', 'Variable 2', 'Correlation']
print("\nHighly Correlated Variables:")
print(high_corr_variables)
```

```python
#3Plot scatter plot between age and maximum heart rate.
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Max HR'], color='green', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.title('Scatter Plot of Age vs. Maximum Heart Rate')
plt.show()
```
